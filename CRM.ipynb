```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame with real estate data
data = {
    'area_type': ['Super built-up Area', 'Plot Area', 'Built-up Area', 'Super built-up Area', 'Super built-up Area'],
    'availability': ['19-Dec', 'Ready To Move', 'Ready To Move', 'Ready To Move', 'Ready To Move'],
    'location': ['Electronic City Phase II', 'Chikka Tirupathi', 'Uttarahalli', 'Lingadheeranahalli', 'Kothanur'],
    'size': ['2 BHK', '4 Bedroom', '3 BHK', '3 BHK', '2 BHK'],
    'society': ['Coomee', 'Theanmp', None, 'Soiewre', None],
    'total_sqft': [1056, 2600, 1440, 1521, 1200],
    'bath': [2, 5, 2, 3, 2],
    'balcony': [1, 3, 3, 1, 1],
    'price': [39.07, 120, 62, 95, 51]
}

df = pd.DataFrame(data)

# Setting up the figure size
plt.rcParams['figure.figsize'] = (30, 12)

# 1. Area Type Distribution
plt.subplot(1, 5, 1)
size = df['area_type'].value_counts()
labels = size.index
colors = ['red', 'green', 'blue', 'cyan']
plt.pie(size, colors=colors, labels=labels, autopct='%1.2f%%')
plt.title('Area Type', fontsize=20)
plt.axis('off')

# 2. Availability Status Distribution
plt.subplot(1, 5, 2)
size = df['availability'].value_counts()
labels = size.index
colors = ['orange', 'purple']
plt.pie(size, colors=colors, labels=labels, autopct='%1.2f%%')
plt.title('Availability Status', fontsize=20)
plt.axis('off')

# 3. Property Size Distribution
plt.subplot(1, 5, 3)
size = df['size'].value_counts()
labels = size.index
colors = ['red', 'green', 'blue', 'cyan', 'orange']
plt.pie(size, colors=colors, labels=labels, autopct='%1.2f%%')
plt.title('Property Size', fontsize=20)
plt.axis('off')

# 4. Location Distribution
plt.subplot(1, 5, 4)
size = df['location'].value_counts()
labels = size.index
colors = ['red', 'green', 'blue', 'cyan', 'grey']
plt.pie(size, colors=colors, labels=labels, autopct='%1.2f%%')
plt.title('Location', fontsize=20)
plt.axis('off')

# 5. Balcony Count Distribution
plt.subplot(1, 5, 5)
size = df['balcony'].value_counts()
labels = size.index
colors = ['red', 'green', 'blue']
plt.pie(size, colors=colors, labels=labels, autopct='%1.2f%%')
plt.title('Balcony Count', fontsize=20)
plt.axis('off')

plt.tight_layout()
plt.grid()
plt.show()
```

Insights
Area Type Distribution:

"Super Built-Up Area" dominates, indicating its popularity among buyers.
Location Popularity:

Most listings are concentrated in Electronic City Phase II, suggesting it is a key area for marketing.
Balcony Count and Price:

Properties with 1 or 3 balconies tend to have distinct price points.
Size Distribution:

2 BHK and 3 BHK properties are the most common, aligning with nuclear family preferences.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame
data = {
    'size': ['2 BHK', '4 Bedroom', '3 BHK', '3 BHK', '2 BHK'],
    'price': [39.07, 120, 62, 95, 51],
    'total_sqft': [1056, 2600, 1440, 1521, 1200]
}

df = pd.DataFrame(data)

# Group by size to calculate averages
size_group = df.groupby('size').mean()

# Extract data for plotting
X = ['Average Price', 'Average Sqft']
two_bhk_scores = [size_group.loc['2 BHK', 'price'], size_group.loc['2 BHK', 'total_sqft']]
three_bhk_scores = [size_group.loc['3 BHK', 'price'], size_group.loc['3 BHK', 'total_sqft']]
four_bedroom_scores = [size_group.loc['4 Bedroom', 'price'], size_group.loc['4 Bedroom', 'total_sqft']]

# Prepare the X-axis
X_axis = np.arange(len(X))

# Plot the data
plt.figure(figsize=(12, 8))
plt.bar(X_axis - 0.3, two_bhk_scores, 0.3, label='2 BHK')
plt.bar(X_axis, three_bhk_scores, 0.3, label='3 BHK')
plt.bar(X_axis + 0.3, four_bedroom_scores, 0.3, label='4 Bedroom')

# Add labels and title
plt.xticks(X_axis, X, fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.title("Comparison of Average Price and Square Footage by Property Size", fontweight='bold', fontsize=14)
plt.legend()

# Display the plot
plt.show()
```

Insights
Average Price:

4 Bedroom properties are priced significantly higher compared to 2 BHK and 3 BHK.
Square Footage:

Larger properties naturally offer more square footage, making them suitable for premium buyers.
Actionable Steps:

For budget-conscious buyers, focus on 2 BHK options.
Premium marketing campaigns should highlight spacious 4 Bedroom offerings.
